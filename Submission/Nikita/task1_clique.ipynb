```python
transaction_class=data['Class'].value_counts()
transaction_class
```

```python
transaction_percent=(transaction_class/284807) * 100
transaction_percent.index = ['0', '1']
transaction_percent
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=transaction_percent.index, y=transaction_percent)
plt.xlabel('Transaction Class')
plt.ylabel('Percentage')
plt.title('Distribution of Fraudulent and Non-Fraudulent Transactions')
plt.show()
```
